 **(10)** Check the ourliers and  handle outliers in peformace score using IQR or percentiles method

```python
# Drop the outlier observations using IQR or Percentiles

q1 = data['performance_score'].quantile(0.25)
q3 = data['performance_score'].quantile(0.75)
iqr = q3-q1

low_lim = q1-1.5*iqr
up_lim = q3+1.5*iqr

outliers = []
for i in data['performance_score']:
  if (i<low_lim) or (i>up_lim):
    outliers.append(i)
print('The performance score which are outliers:', outliers)

performance_outlier = (data['performance_score']<low_lim) | (data['performance_score']>up_lim)
data.loc[performance_outlier].index
```

**(11)** check the gender column and do Dummy encoding

```python
gender_encoded = pd.get_dummies(data['gender'])
gender_encoded.columns = ['Female', 'Male']
gender_encoded
```

```python
data = data.drop(['gender'], axis=1)
```

```python
data = pd.concat([data, gender_encoded], axis=1)
data
```
